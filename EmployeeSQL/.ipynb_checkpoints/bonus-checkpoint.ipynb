```python
# import dependencies
from sqlalchemy import create_engine
import config
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# establish connection
database_url = f"postgresql://user:password@localhost:5432/employee_db"
engine = create_engine(database_url)
connection = engine.connect()
```

```python
# create an employees dataframe
emp_df = pd.read_sql("select * from employees", connection)
emp_df.fillna("NaN")
```

```python
# create a histogram for the salaries
salaries_df = pd.read_sql("select * from salaries", connection)
salaries_df.fillna("NaN")
```

```python
plt.hist(salaries_df["salary"], range = (40000, salaries_df['salary'].max()), color = 'red')

plt.title("Most Common Salary Ranges")

plt.xlabel("salary range")

plt.ylabel("frequency")

plt.show()
```

```python
# create a bar chart that shows average pay by position
titles_df = pd.read_sql("select * from titles", connection)
titles_df.fillna("NaN")

# rename title_id to emp_title
titles_df = titles_df.rename(columns = {"title_id": "emp_title"})
titles_df
```

```python
# merge employee and salary dataframes
merge_titles = pd.merge(emp_df, salaries_df, on = "emp_no")
merge_titles.head()
```

```python
avg_salary = merge_titles.groupby(["emp_title"])["salary"].mean()
avg_salary
```

```python
sal_titles = pd.merge(titles_df, avg_salary, on = "emp_title")
sal_titles
```
